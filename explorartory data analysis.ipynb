# EXPLORATORY DATA ANALYSIS

```python
df.head()
```

### Top 5 Cities Sales

```python
top_5_cities=(df.groupby("City")["Sales"].sum().nlargest(5).reset_index())
top_5_cities
```

```python
plt.figure(figsize=(8,5))
ax = sns.barplot( x= "City", y ="Sales", data=top_5_cities, palette="Blues")

for container in ax.containers:
    ax.bar_label(container,fmt="%.0f")



plt.title(" top 5 cities sales")
plt.xlabel("CITY")
plt.ylabel("Sales")
plt.show()
```

### REGIONWISE SALES


```python
Sales_Regionwise= (df.groupby("Region")["Sales"].sum().reset_index().sort_values(by="Sales",ascending=False))
```

```python
Sales_Regionwise
```

```python
plt.figure(figsize=(8,5))
ax=sns.barplot(x="Region",y="Sales",data=Sales_Regionwise,palette="Blues")
for container in ax.containers:
    ax.bar_label(container)
plt.show()
```

```python
ax=sns.countplot(x="Sub-Category",data=df)
sns.set(rc={'figure.figsize':(25,12)})
for i in ax.containers:
    ax.bar_label(i)
plt.title("Distribution of Customer Purchases Across Product Sub-Categories")
plt.xlabel("sub category")


plt.show()
```

### Top 5 Product Sales

```python
Product = df.groupby("Product Name")["Sales"].sum().nlargest(5).reset_index()
```

```python
Product
```

```python
plt.figure(figsize=(30,10))
ax = sns.barplot( x= "Product Name", y ="Sales", data=Product, palette="Blues")

for container in ax.containers:
    ax.bar_label(container,fmt="%.0f")



plt.title(" top product cities sales")
plt.xlabel("Product")
plt.ylabel("Sales")
plt.show()
```

### TOTAL SALE PER YEAR 

```python
df["Order Date"]=pd.to_datetime(df["Order Date"],dayfirst=True, errors="coerce")
df["year"]=df["Order Date"].dt.year
```

```python
Sale_Per_Year=df.groupby("year")["Sales"].sum().reset_index()
```

```python
Sale_Per_Year
```

```python
plt.figure(figsize=(15,10))
ax = sns.barplot( x= "year", y ="Sales", data=Sale_Per_Year, palette="Blues")

for container in ax.containers:
    ax.bar_label(container,fmt="%.0f")



plt.title(" top product cities sales")
plt.xlabel("Product")
plt.ylabel("Sales")
plt.show()
```

# CONCLUSION

### Conclusion:
The analysis highlights that New York City recorded the highest sales overall, with the West region emerging as the top-performing region. Within product categories, the Binder sub-category generated the maximum sales, while at the product level, the Canon imageCLASS 2200 Advanced Copier was the highest-selling item. Year-wise analysis shows that the peak sales occurred in 2018, marking it as the most successful year in terms of revenue.
